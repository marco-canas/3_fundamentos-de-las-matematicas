# Clasificación de funciones

Clasifique cada función como función potencia, función raíz,
polinomio (señale su grado), función racional, función algebraica, función trigonométrica, función exponencial o función logarítmica.

1. $ f(x) = \sqrt{x} $

Se llama función raiz cuadrada


```python
import numpy as np 

from matplotlib import pyplot as plt 

plt.figure(figsize = (10,6))

plt.title('Gráfico de la función raiz es una semiparábola')

X = np.linspace(0,10, 11)
F = np.sqrt(X)

plt.plot(X, F, 'ro')
plt.plot(X, F, 'b-', label = '$f(x) = \sqrt{x}$')

plt.grid(alpha = 0.4)

plt.xlabel('abscisas $X$')
plt.ylabel('Ordenadas $F$')

plt.legend() 

plt.savefig('grafico_funcion_raiz.jpg')

plt.show()  
```
