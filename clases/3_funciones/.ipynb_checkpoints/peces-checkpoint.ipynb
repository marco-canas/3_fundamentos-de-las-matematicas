### Población de peces 

La población de peces de cierto lago sube y baja de acuerdo con la fórmula:

$$ F(t) = 1000(30 + 17t – t^{2}) $$

Aquí $F$ es el número de peces en el tiempo $t$, donde $t$ se mide en años desde el 1 de enero de 2002, cuando la población de peces se estimó por primera vez.  

(a) ¿En qué fecha la población de peces será otra vez la misma de como era el 1 de enero de 2002?

(b) ¿Antes de qué fecha habrán muerto todos los peces del lago?

## Solución

## Identificación de conceptos y procedimientos asociados a la solución del problema

* Función cuadrática $f(x) = ax^{2} + bx + c$
* Gráfico de una función cuadrática: parábola.
* Concavidad de una función cuadrática depende del coeficiente $a$

```python
import numpy as np 
from matplotlib import pyplot as plt 

plt.figure(figsize = (13,5))

plt.title('Modelación gráfica de la dinamica poblacional de peces')  

plt.xlabel('Tiempo $t$')
plt.ylabel('Población de peces $F$') # fish es pez en inglés

a,b = 0, 19 # dominio de graficación

f = lambda t: 1_000*(30 + 17*t - t**2) 
plt.axis([a,b, 0, 110_000])

T = np.linspace(a,b, 30)
F = np.array([f(t) for t in T]) 

plt.plot(T, F, 'b-', label = '$F(t) = 10000(30 + 17t - t^{2})$')

plt.xticks(np.arange(a,b))
plt.yticks(np.arange(0,110_000, 10_000))

plt.plot([a, b], [0,0], 'g--') # trazar el eje horizontal 
plt.grid(alpha = 0.4)

plt.scatter([17],[30_000], color = 'red' )

plt.plot([a, 17], [30_000, 30_000], 'r--') # línea punteada a una altura de 30_000
plt.plot([17, 17], [0, 30_000], 'r--') # línea punteada vertical a una posición t = 17


plt.legend()     

plt.savefig('modelacion_peces.jpg')

plt.show() 
```

## Interpretación del gráfico 

* La población inicial es de 30000 peces lo cual se evidencia en el intersecto del gráfico con el eje $F$. 

* La población crece en los primeros 8.5 años y decrece a cero en el intervalo de 8.5 a 18.6 años. 

* la población máxima es de aproximadamente 110000 y se alcanza a los 8.5. 

* Al cabo de 17 años la población vuelve a ser la población inicial 

## Resolver las preguntas del problema pero con sympy

El modelo que sesuelve la primera pregunta ((a) ¿En qué fecha la población de peces será otra vez la misma de como era el 1 de enero de 2002?) es:

$$ f(t) = f(0) $$




```python
from sympy import * 
t,F = var('t,F')   # definir variables 
F = Function('F')
# (a) ¿En qué fecha la población de peces será otra vez la misma de como era el 1 de enero de 2002?
f = lambda t: 1000*(30 + 17*t - t**2) 
eq_1 = Eq(f(t), f(0))     # modelo que resuelve la primera pregunta. 
sol_eq_1 = solve(eq_1, t)
sol_eq_1[1]
```

Es decir, en dos momentos del tiempo, la población es la población inicial. Luego, a los 17 años la población vuelve a ser la población inicial. 

## Modelo que resuelve la segunda pregunta ((b) ¿Antes de qué fecha habrán muerto todos los peces del lago?):

$$ f(t) = 0 $$


```python
# solución a la segunda pregunta usando sympy
# (b) ¿Antes de qué fecha habrán muerto todos los peces del lago?
eq_2 = Eq(f(t), 0)     # modelo 2 
sol_eq_2 = solve(eq_2, t)
sol_eq_2[0].evalf(), sol_eq_2[1].evalf()
```

Por lo tanto, al cabo de 18.61 años, la población de peces se extingue. 

# Experimentación con deslizador para cambiar la concavidad de una parábola

```python
import numpy as np 

from matplotlib import pyplot as plt 

from ipywidgets import interact 



def graficar_parabola(a=1, b=1, c=1):  # f(t) =  at^2 + bt + c
    a_,b_ = -4,5 # defino dominio de graficación
    T = np.arange(a_,b_, 0.4) 
    def funcion_cuadratica(t):
        f_ = a*t**2 + b*t + c
        return f_
    F = np.array([funcion_cuadratica(t) for t in T])
    plt.plot(T, F)
    plt.axis([a_, b_, -20,20])
    plt.grid(alpha = 0.4)

interact(graficar_parabola, a = (-1,1,2), b = (-1,1,2), c = (-1,1,2))
```

## Interpretación de la experimentación interactivas con deslizadores

* El parámetro que cambia la concavidad de la parábola es el coeficiente $a$ del término cuadrático en la parábola. 

* El parámetro b conttrola desplazamientos horizontales.

* El parámetro c controla movoimientos verticales de la parábola.
