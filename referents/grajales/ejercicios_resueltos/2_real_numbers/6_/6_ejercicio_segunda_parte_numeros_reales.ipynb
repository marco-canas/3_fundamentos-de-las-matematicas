6. Halle $x$ de manera que la distancia entre los dos puntos sea 5.  

a) $(0; 0), (x; -4)$  

b) $(2; 0), (x; 2)$

Solución:

$$ dist(A, B) = \sqrt{(Ax - Bx)^{2} + (Ay-By)^{2}} $$

$$ 5 = \sqrt{(0- x)^{2} + (0-(-4))^{2}} $$

$$ \Rightarrow \quad 5^{2} = \left( \sqrt{(0 - x)^{2} + (0-(-4))^{2}} \right)^{2} $$

$$ \Rightarrow \quad 25 = (0 - x)^{2} + (0-(-4))^{2} $$

$$ 25 = x^{2} + 16 $$

$$ 25 - 16 = x^{2} $$

$$ 9 = x^{2} $$

$$ \pm 3 = x $$
