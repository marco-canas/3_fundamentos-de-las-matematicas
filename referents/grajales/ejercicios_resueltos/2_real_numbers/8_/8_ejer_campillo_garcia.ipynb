```python
import matplotlib.pyplot as plt

# Coordenadas del primer punto
x1, y1 = 2/3, 1/3

# Coordenadas del segundo punto
x2, y2 = 5/6, 1

# Crear una figura y un conjunto de ejes
fig, ax = plt.subplots()

# Graficar los puntos
ax.plot(x1, y1, 'ro', label='Punto 1')
ax.plot(x2, y2, 'bo', label='Punto 2')

# Etiquetar los puntos
ax.annotate(f'({x1}, {y1})', (x1, y1), textcoords="offset points", xytext=(-20,10), ha='center')
ax.annotate(f'({x2}, {y2})', (x2, y2), textcoords="offset points", xytext=(20,10), ha='center')

# Agregar leyenda
ax.legend()

# Mostrar la gráfica
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Gráfico de dos puntos')
plt.grid(True)
plt.show()


```

```python
from sympy import Point, Rational

# Definir los puntos (x1, y1) y (x2, y2)
x1, y1 = Rational(2, 3), Rational(1, 3)
x2, y2 = Rational(5, 6), Rational(1, 1)

# Crear los objetos Point
punto1 = Point(x1, y1)
punto2 = Point(x2, y2)

# Calcular el punto medio
punto_medio = punto1.midpoint(punto2)

# Obtener las coordenadas del punto medio
x_medio, y_medio = punto_medio

# Imprimir el resultado
print(f"El punto medio del segmento es ({x_medio}, {y_medio})")
```
