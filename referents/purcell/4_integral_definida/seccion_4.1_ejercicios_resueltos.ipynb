### Ejercicio 53 

$$ y = x + 2, \quad \quad [0,1] = [a,b] $$

## Empezamos graficandoo la región de integración

```python
import numpy as np 
import matplotlib.pyplot as plt 

dominio = np.arange(-4,5)

def f(x):
    return x + 2  # función lineal de la forma y = mx + b

rango = f(dominio)

plt.plot(dominio, rango, label = r'$f(x) = x + 2$')  # gráfico de la función lineal $f(x) = x + 2$
plt.plot([-4,4], [0,0], color = 'g')  # el eje horizontal o eje x
plt.plot([0,0], [-2,6], color = 'g')  # el eje vertical o eje y
plt.title('Gráfico de una región para determinar su área')
plt.xlabel('eje x')
plt.ylabel('eje y')
plt.legend() 
plt.grid()
# Gráfico de la región de integración

plt.plot([0, 1],[2, 3], 'r--', linewidth = 2)
plt.plot([0, 0],[0, 2], 'r--', linewidth = 2)
plt.plot([1, 1],[0, 3], 'r--', linewidth = 2)
plt.plot([0, 1],[0, 0], 'r--', linewidth = 2)
plt.show()

```

# Construcción de una función de Python que grafique la región cuya área es el área bajo la curva

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función lambda
f = lambda x: x**2  # Cambia esto por tu función

# Definir los límites del intervalo cerrado
a = -1  # Límite inferior
b = 2   # Límite superior

# Generar puntos para graficar la función y las líneas verticales
x_vals = np.linspace(a, b, 400)
y_vals = f(x_vals)
x_a = np.array([a, a])
y_a = np.array([0, f(a)])
x_b = np.array([b, b])
y_b = np.array([0, f(b)])

# Graficar la función y las líneas verticales
plt.plot(x_vals, y_vals, label='y = f(x)')
plt.plot(x_a, y_a, 'r--', label=f'x = {a}')
plt.plot(x_b, y_b, 'g--', label=f'x = {b}')
plt.fill_between(x_vals, y_vals, alpha=0.2)

# Etiquetas y leyenda
plt.xlabel('x')
plt.ylabel('y')
plt.title('Región de Integración')
plt.legend()

# Mostrar el gráfico
plt.grid()
plt.show()
```
